```python
smile = pd.read_csv('SmartVision Smiles_Find Dialogue_Table.csv')
smile.head(9)
```

```python
corpus =  smile[smile.Success == 0]
corpus.head(10)
```

```python
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)

stemmer = PorterStemmer()

def stopword_and_stem_tokenizer(corpus):
    """
    Custom tokenizer function to be passed in TfidfVectorizer.
    Ignores english stopwords and applies stemming to remaining tokens.
    """
    tokens = [word for sentence in nltk.word_tokenize(corpus) for word in nltk.word_tokenize(sentence)]
    valid_tokens = []
    for token in tokens:
      #  if re.search('[a-zA-Z]', token): # ignore non-alphabetic characters
        valid_tokens.append(token)
    result = [t for t in valid_tokens if t not in [stopwords.words('greek'),stopwords.words('english')]]
    return result
```

```python
ss = np.array(corpus['Chat ID'])
sm = np.array(corpus['Question'])
si = np.array(corpus['Success'])
sl = np.array(corpus['Timestamp'])
se = np.array(corpus['Language'])
stopwords_acc = string.punctuation
```

```python

corpora = []
for i,sentence in enumerate(sm):
    st = stopword_and_stem_tokenizer(sentence)
    stl = str(st).replace("'","").replace('[','').replace(']','').replace(',','').replace('?','')
    curr = word_tokenize(stl)
    curr = [word for word in curr if word not in stopwords_acc]
    corpora.append(curr)
    
```

```python
class CS_D(object):
    def __init__(self, data,texts, amp,tea,depth=1):
        super(object, self).__init__()
        self.data = data
        self.texts = texts
        self.amp = amp
        self.depth = depth
        self.it = tea
    
    
    def _42_(self,en):
        l2 = []
        for i in en:
            l3 = []
            for j in i:
                k = j.split(' ')
                l = ' '.join(k[:-1])
                u = ' '.join(k[1:])
                l3.append([l,u])
            l2.append(l3)
        
        gl = []
        for x,i in enumerate(l2):
            l1 = []
            for y,j in enumerate(i):
                z = 0
                dl = []
                for k in i:
                    z += 1
                    if j[0]==k[1]: 
                        key = j[1].split(' ')[-1:]
                        v = en[x][z-1] +' '+ key[0]
                        dl.append(v)
                        
                if dl != []: l1.append(dl);
            if l1 != []: gl.append(l1);
                
        return gl[0]
    
    def _777_(self,en,gr,engr):
        good = []
        for x,i in enumerate(en):
            l1 = []
            for y,j in enumerate(gr):
                z = 0
                dl = []
                for k in i:
                    z += 1
                    if j[0]==k[1]: 
                        key = j[1].split(' ')[-1:]
                        v = engr[x][z-1] +' '+ key[0]
                        dl.append(v)
                        
                if dl != []: l1.append(dl);
            if l1 != []: 
                l = [item for sub_list in l1 for item in sub_list]
                good.append(l)
        return good
    
    
    def _64_(self,en):
        l2 = []
        for i in en:
            k = i.split(' ')
            l = ' '.join(k[:-1])
            u = ' '.join(k[1:])
            l2.append([l,u])
        return l2
                
    
    def _6_(self,f,g):
        print(f)
        f_y = []
        for i in f:
            y = self._64_(i)
            f_y.append(y)

        g_x = []
        for i in g:
            x = self._64_([i])[0]
            g_x.append(x)
        
        f_g = self._777_(f_y,g_x,f)
        
        return f_g
        
    def make_n_grams(self,data,depth):
        oui = []
        n = self.amp+depth
        for i in range(1,n):
            non = Counter()
            for sent in data:
                non.update(
                    [gram for gram in ngrams(
                        sent,
                        i+1, 
                        pad_left=False, 
                        pad_right=False)
                    ]
                )
            oui.append(non)
        return oui
    
    def find(self,data,depth):
        csis = self.make_n_grams(data,depth)
        n = self.amp+depth
        clusters = []
        for i in range(2,n):
            found = [' '.join(key) for key, _ in csis[i-1].most_common(self.it)]
            clusters.append(found)
        return clusters
    

    def cluster(self,data):
        depth = self.depth
        print(depth)
        for d in range(depth):
            if d == 0:
                kyp = self.find(data,d)
                print(kyp[0])
                meaning = self._42_(kyp)
               # print(kyp[0],'\n\n')
            else:
                print()
                kp = self.find(data,d)[d]
                meaning = self._6_(meaning,kp)
                print()
        print(meaning)
        n = self.amp
        kept = self.texts.copy()
        l_list = []
        for i in range(1,n-1):
            for j in kyp[i-1]:
                c_list = []
                for k in kept:
                    if j in k:
                        c_list.append(k)
                        kept.remove(k)
                l_list.append(c_list)
        return l_list

    
```
